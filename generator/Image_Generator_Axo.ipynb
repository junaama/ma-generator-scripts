```python
# Generate all features

TOTAL_AXOLOTLS = 10000

features = []

def createCombination():
    """
    Creates all possible combinations of features
    """
    feature = {}

    feature["Backgrounds"] = random.choices(backgrounds, backgroundweights)[0]
    feature["Skins"] = random.choices(skin, skinweights)[0]
    feature["Eyes"] = random.choices(eyes, eyesweights)[0]
    feature["Mouths"] = random.choices(mouth, mouthweights)[0]

    feature["Headwears"] = random.choices(headwears, headwearweights)[0]
    feature["Drapery"] = random.choices(drapery, draperyweights)[0]
    feature["Vest"] = random.choices(vest, vestweights)[0]
    feature["Smudges"] = random.choices(smudges, smudgesweights)[0]
    

    if feature["Headwears"] == "":
        del feature["Headwears"]
    if feature["Drapery"] == "":
        del feature["Drapery"]
    if feature["Vest"] == "":
        del feature["Vest"]
    if feature["Smudges"] == "":
        del feature["Smudges"]
    
    if feature in features:
        return createCombination()
    else:
        return feature 

for i in range(TOTAL_AXOLOTLS):
    newfeaturecombination = createCombination()
    
    features.append(newfeaturecombination)
```

```python
def isUnique(x):
    """
    Returns boolean of if each combination of features is unique.
    """
    seen = list()
    return not any(i in seen or seen.append(i) for i in x)
print(isUnique(features))
```

```python
# Create a token ID and add to each object in features
i = 0
for item in features:
    item["tokenId"] = i 
    i = i + 1
```

```python
# Generate Images

for item in features:
    imagebackground = Image.open(f'./7 BACKGROUND/{backgroundfiles[item["Backgrounds"]]}.png').convert('RGBA')
    imageskin = Image.open(f'./6 SKIN/{skinfiles[item["Skins"]]}.png').convert('RGBA')
    imagemouth = Image.open(f'./2 MOUTH/{mouthfiles[item["Mouths"]]}.png').convert('RGBA')
    imageeye = Image.open(f'./1 EYES/{eyesfiles[item["Eyes"]]}.png').convert('RGBA')
    
    if "Headwears" in item:
        imageheadwears = Image.open(f'./0 HEADWEAR/{headwearfiles[item["Headwears"]]}.png').convert('RGBA')
    else:
        imageheadwears = False
    if "Drapery" in item:
        imagedrapery = Image.open(f'./4 DRAPERY/{draperyfiles[item["Drapery"]]}.png').convert('RGBA')
    else:
        imagedrapery = False
    if "Vest" in item:    
        imagevest = Image.open(f'./3 VEST/{vestfiles[item["Vest"]]}.png').convert('RGBA')
    else:
        imagevest = False
    if "Smudges" in item: 
        imagesmudges = Image.open(f'./5 SMUDGES/{smudgesfiles[item["Smudges"]]}.png').convert('RGBA')
    else:
        imagesmudges = False

    # Create composite image
    
    com1 = Image.alpha_composite(imagebackground, imageskin)
    if imagesmudges:
        com1 = Image.alpha_composite(com1, imagesmudges)
    if imagedrapery:
        com1 = Image.alpha_composite(com1, imagedrapery)
    if imagevest:
        com1 = Image.alpha_composite(com1, imagevest)

    if item["Mouths"] == "Gum" or item["Mouths"] == "Tongue" or item["Mouths"] == "Chupa" or item["Mouths"] == "Covid Mask":
        com2 = Image.alpha_composite(com1, imagemouth)
        com3 = Image.alpha_composite(com2, imageeye)
    else:
        com2 = Image.alpha_composite(com1, imageeye)
        com3 = Image.alpha_composite(com2, imagemouth)

    

    endcom = com3
    # if imagedrapery:
    #     endcom = Image.alpha_composite(endcom, imagedrapery)
    # if imagesmudges:
    #     endcom = Image.alpha_composite(endcom, imagesmudges)    
    if imageheadwears:
        endcom = Image.alpha_composite(endcom, imageheadwears)
    # if imagevest:
    #     endcom = Image.alpha_composite(endcom, imagevest)

    if endcom:
        finalcom = Image.alpha_composite(com3, endcom)
    else:
        finalcom = com3

    rgb_image = finalcom.convert('RGB')
    filename = str(item["tokenId"]) + ".png"
    rgb_image.save("./output/" + filename)
    print(f'{str(item["tokenId"])} done')
```
